```python
#Use a function to get number of men on base at any given time
def men_on_base(df):
    df['on_1b'] = ~df['on_1b'].isnull()
    df['on_2b'] = ~df['on_2b'].isnull()
    df['on_3b'] = ~df['on_3b'].isnull()
    df['men_on'] = df[['on_1b','on_2b','on_3b']].sum(axis=1)
```

```python
data = statcast(start_dt='2018-03-20',end_dt='2019-10-01')

#data = pd.read_pickle('statcast.pkl')

#Add our 'men on base' ASAP so we don't have to do it in several places later
men_on_base(data)

#We don't have a consistent "batting team" or "pitching team". Adding those fields now.

data['batting_team'] = np.where(data['inning_topbot'] =='Top', data['away_team'], data['home_team'])
data['pitching_team']= np.where(data['inning_topbot'] =='Bot', data['away_team'], data['home_team'])

#Now get batting score and pitching score
data['batting_score'] = np.where(data['inning_topbot'] =='Top', data['away_score'], data['home_score'])
data['pitching_score'] = np.where(data['inning_topbot']=='Bot', data['away_score'], data['home_score'])

#Uncomment the line below when you wish to pickle the data.
#data.to_pickle('statcast.pkl')
```

We see that there are batted balls in every category that are misclassified. It is known that Statcast isn't perfect, but we also see that it gets it right most of the time - that's what we expect. In this analysis I'm making the assumption that the launch angle is correct and redefining the batted ball type based on that. 

I am inclined to go this way because for many observations with a valid launch angle, there is a NaN missing value for the batted ball type. I don't know how Statcast fills these values, so I'm not at liberty to say what caused this discrepancy in the data.

http://m.mlb.com/glossary/statcast/launch-angle

```python
#Define a function to reset the batted ball types
def batted_ball(launch_angle):
    if launch_angle < 10:
        return 'ground_ball'
    elif launch_angle < 25:
        return 'line_drive'
    elif launch_angle <50:
        return 'fly_ball'
    else:
        return 'pop_up'

#Get only data for pitches where there is a corresponding launch angle
angle_eligible = data[np.isfinite(data['launch_angle'])]

#Apply the function
angle_eligible['bb_type'] = angle_eligible['launch_angle'].apply(batted_ball)

ax = sns.catplot(x="bb_type", y="launch_angle"
            ,data=angle_eligible[['launch_angle','bb_type']].dropna().reset_index()
            ,order=['ground_ball','line_drive','fly_ball','pop_up'])
plt.title('Amended Batted Ball Type by Launch Angle')
plt.xlabel('Amended Batted Ball Type')
plt.ylabel('Launch Angle')
```

```python
#Coerce different on-field events into a few distinct buckets
def event_creator(event):
    if event in (['field_out','grounded_into_double_play','force_out','double_play','fielders_choice_out','fielders_choice'
                  ,'other_out','triple_play']):
        return 'out'
    elif event in (['hit_by_pitch','strikeout','catcher_interf','walk','pickoff_1b','caught_Stealing_2b','pickoff_caught_stealing_2b'
                    ,'caught_stealing_3b','batter_interference','pickoff_2b','run','caught_stealing_home','pickoff_caught_stealing_3b'
                    ,'field_error','caught_stealing_2b']):
        return 'other' #strikeout is not 'out' because outs only look at balls in play
    elif event in (['home_run']):
        return 'home_run'
    elif event in (['single']):
        return 'single'
    elif event in (['double']):
        return 'double'
    elif event in (['triple']):
        return 'triple'
    elif event in (['sac_fly','sac_bunt','sac_fly_double_play','sac_bunt_double_play']):
        return 'sacrifice'
    else:
        return 'unexpected_play_type'
play_type = angle_eligible.dropna(subset=['events'])
play_type['event_type'] = play_type['events'].apply(event_creator)
play_type['event_type'].unique()

#Visualize 
ax = sns.violinplot(x="events", y="hit_distance_sc"
            ,data=play_type[['events','hit_distance_sc','event_type']].query('event_type == "out"'))
sns.set(rc={'figure.figsize':(14.7,14.27)})
plt.savefig('outs_violin.png')

ax_playtype = sns.violinplot(x='event_type', y='hit_distance_sc'
                            ,data=play_type[['event_type','hit_distance_sc']])
plt.savefig('playtype_violin.png')
```

So let's check - is launch angle statistically correlated with hit distance?

Though not explicitly covered here, I used normaltest() from scipy.stats, which showed a non-normal distribution for both hit distance and launch angle. This might be expected - there are a lot more ground balls than fly balls or line drives.

My research indicates that a violation of the normality assumption may not have a significant effect on Pearson correlation. I have chosen to use Kendall's Tau as a correlation measure (concordant pairs) as it allows for non-normal data and non-linear relationships by analyzing concordant pairs. I am open to feedback on this, so please feel free to send a message any time.

Pearson correlation is insensitive to normality violations: https://journals.sagepub.com/doi/abs/10.2466/pms.1976.43.3f.1319?journalCode=pmsb

Further reading on Kendall's Tau: https://www.statisticssolutions.com/kendalls-tau-and-spearmans-rank-correlation-coefficient/



```python
#Get only data for which there is both a measured hit distance and launch angle (.isfinite() weeds out NaN values)
launch_angle = data[np.isfinite(data['hit_distance_sc'])][np.isfinite(data['launch_angle'])][['hit_distance_sc','launch_angle']].reset_index().drop(['index'], axis=1) 

#Create new variables for ranks in hit distance and launch angle
launch_angle['dist_rank'] = launch_angle['hit_distance_sc'].rank()
launch_angle['angle_rank'] = launch_angle['launch_angle'].rank()

#Run a Kendall's Tau test for correlation between concordant/discordant pairs
kendalltau(launch_angle['dist_rank'], launch_angle['angle_rank'])
```

The null hypothesis for a Kendall's Tau test for correlation for concordant/discordant pairs is that there is an absence of a relationship. We see a moderate relationship (**tau = 0.5572**) with a p-value close to zero, showing that there *is* a relationship between hit distance and launch angle, which goes with the prevailing theory.

By definition, the pd.rank() function does not allow for ties (and if it did we would have a different problem on our hands), so we may have an artificially increased number of discordant pairs, driving down the perceived correlation from the test. This is important to note because we may actually have a stronger relationship than the test might tell us.

Next we will take a look at the same information while substituting launch speed instead of launch angle.

```python
ax = sns.scatterplot(x='hit_distance_sc', y='launch_speed'
                     ,data=data.query('events == "home_run"'))
plt.title('2018/2019 Home Runs - Hit Distance by Launch Speed')
plt.xlabel('Hit Distance (Feet)')
plt.ylabel('Launch Speed')
```

```python
#Get only data for which there is both a measured hit distance and launch angle (.isfinite() weeds out NaN values)
launch_speed = data[np.isfinite(data['hit_distance_sc'])][np.isfinite(data['launch_speed'])][['hit_distance_sc','launch_speed']].reset_index().drop(['index'], axis=1) 

#Create new variables for ranks in hit distance and launch angle
launch_speed['dist_rank'] = launch_speed['hit_distance_sc'].rank()
launch_speed['speed_rank'] = launch_speed['launch_speed'].rank()

#Run a Kendall's Tau test for correlation between concordant/discordant pairs
print(kendalltau(launch_speed['dist_rank'], launch_speed['speed_rank']))
print(f"Pearson R result: {pearsonr(launch_speed['hit_distance_sc'], launch_speed['launch_speed'])}")
```

Here we observe a mild correlation between launch speed and hit distance, though roughly half of what was observed for launch angle. It looks like we may have a linear correlation between the two variables, so I checked Pearson correlation, and yet we fail to see a strong linear correlation there either - a good reminder that the numbers don't lie.

**Now we turn our attention toward how many additional runs were created because of the long ball.**

```python
#Get a final run score
total_runs2019 = data[['batting_team','batting_score','game_date']].query('game_date > "2019-01-01"')
total_runs2019 = total_runs2019.groupby(['batting_team','game_date']).max().reset_index()
total_runs2019 = total_runs2019.groupby('batting_team').sum().reset_index()
total_runs2019.columns = ['team','runs']

total_runs2018 = data[['batting_team','batting_score','game_date']].query('game_date < "2019-01-01"')
total_runs2018 = total_runs2018.groupby(['batting_team','game_date']).max().reset_index()
total_runs2018 = total_runs2018.groupby('batting_team').sum().reset_index()
total_runs2018.columns = ['team','runs']

total_runs_allowed2019 = data[['pitching_team','batting_score','game_date']].query('game_date > "2019-01-01"')
total_runs_allowed2019 = total_runs_allowed2019.groupby(['pitching_team','game_date']).max().reset_index()
total_runs_allowed2019 = total_runs_allowed2019.groupby('pitching_team').sum().reset_index()
total_runs_allowed2019.columns = ['team','runs_allowed']

total_runs_allowed2018 = data[['pitching_team','batting_score','game_date']].query('game_date < "2019-01-01"')
total_runs_allowed2018 = total_runs_allowed2018.groupby(['pitching_team','game_date']).max().reset_index()
total_runs_allowed2018 = total_runs_allowed2018.groupby('pitching_team').sum().reset_index()
total_runs_allowed2018.columns = ['team','runs_allowed']

run_summary2019 = pd.merge(total_runs2019, total_runs_allowed2019, on='team')
run_summary2019.columns = ['team','runs2019','runs_allowed2019']
run_summary2018 = pd.merge(total_runs2018, total_runs_allowed2018, on='team')
run_summary2018.columns = ['team','runs2018','runs_allowed2018']
run_summary = pd.merge(run_summary2019, run_summary2018, on='team')
```

Let's take the Minnesota Twins as an example - they hit 85% more home runs in 2019 as they did in 2018, and 55% of their runs came strictly from going yard. If they had a modest, league-average 20% increase in home runs, what would their season have looked like?


*Because baseball is a strategic game, pitchers, hitters, runners, and managers adapt their strategies to the specific situation. This makes it difficult to retroactively adjust. With this as a known limitation, I cautiously proceed with the extrapolation below.*

The Twins had 514 runs from 307 home runs, equating to 1.67 runs per game from their collective moonshots. To bring them back to a league-average 20% rise in home runs, we will take away 108 home runs ((1- (20/85)) * 141). At the rate of 1.674 runs per dinger, this takes away 180 runs from their season, or 1.11 runs per game.

```python
#Selected team scores
selected_team = input('What team would you like to examine? ').upper()
team_data = data.query("game_date > '2019-01-01'")
team_data = team_data[['game_date','game_pk','home_team','post_home_score','post_away_score','away_team']].query(f'away_team == "{selected_team}" or home_team == "{selected_team}"').groupby(['game_pk']).max()
team_data['diff'] = team_data['post_home_score'] - team_data['post_away_score']
```

```python
#Conditional assignment using np.where
team_data['verdict'] = np.where(((team_data['away_team'] == f"{selected_team}") & (team_data['diff'] < 0)) | ((team_data['home_team'] == f"{selected_team}") & (team_data['diff'] > 0)), 'W', 'L')
```

```python
close_games = team_data.query('abs(diff) < 3 & verdict == "W"')
one_run_wins = close_games.query('abs(diff) == 1').shape[0]
two_run_wins = close_games.query('abs(diff) == 2').shape[0]
print(one_run_wins)
print(two_run_wins)
```

In the example I'm running (with Minnesota Twins, MIN, as selected team), since the Twins lose 1.11 runs per game, they will lose 8 of every 9 one-run games, and lose 1 of every 9 two-run games that they had previously won.

This means they lose an additional 17 games on one-run games and 2 two-run games. The 2019 Twins record is now 82-80.
